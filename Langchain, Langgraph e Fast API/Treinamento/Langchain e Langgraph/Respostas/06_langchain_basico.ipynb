```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get("OPENAI_API_KEY")

import os
from langchain_openai import ChatOpenAI
```

```python
def mandarPergunta(pergunta):
  configLLM = ChatOpenAI(model="gpt-3.5-turbo", temperature= 0.5, api_key= OPENAI_API_KEY)
  resposta = configLLM.invoke(pergunta)
  return resposta.content
```

```python
pergunta = input("Converse comigo: ")
mandarPergunta(pergunta)
```

# Trabalhando com mensagens

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage


configLLM = ChatOpenAI(
    model= "gpt-3.5-turbo",
    temperature= 0.7,
    api_key= OPENAI_API_KEY
)

while True:
  perguntaUser = input("Converse comigo: ")

  if "Parar conversa" in perguntaUser:
    print("Conversa finalizada")
    break

  else:
    mensagens = [
        SystemMessage("Voce é um assistente focado em culinaria"),
        HumanMessage(perguntaUser)
    ]

    resposta = configLLM.invoke(mensagens)
    print(resposta.content)
```
